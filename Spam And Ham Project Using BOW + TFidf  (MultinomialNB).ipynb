```python
import pandas as pd
df = pd.read_csv("spam.csv",encoding="latin1")
df.head()
```

```python
df.columns
```

```python
# Rename 'v1' to 'label' and 'v2' to 'Message'
df.rename(columns={'v1': 'label', 'v2': 'Message'}, inplace=True)

# Drop the unwanted columns
df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], inplace=True)

# Display the updated DataFrame
print(df.head())
```

```python
df.head()
```

```python
df.info()
```

### Data Cleaning and Preprocessing

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lz = WordNetLemmatizer()
from nltk.stem import SnowballStemmer
sb = SnowballStemmer('english')
from nltk.stem import PorterStemmer
pt = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(df)):
    review = re.sub('[^a-zA-Z]',' ',df['Message'][i])
    review = review.lower()
    review =review.split()
    review = [lz.lemmatize(word)for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
### Output Feature
y = pd.get_dummies(df['label'])
y = y.iloc[:,0].values
```

```python
y.shape
```

### Train_Test_split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size=0.25, random_state=42)
```

### Create Bag of Words model

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 2500 , ngram_range = (2,2))
```

```python
len(X_train),len(y_train)
```

```python
X_train
```

```python
## Independent feature
X_train = cv.fit_transform(X_train).toarray()
```

```python
X_train
```

```python
X_test = cv.transform(X_test).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems = 30,linewidth = 100000,formatter = dict(float = lambda x:"%.3g" %x))
X_train
```

```python
cv.vocabulary_
```

### Machine Learning Model Training (Multinomial Naive Bayes Algorithm)

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train,y_train)
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

### Create TFIDF model

```python
### Output Feature
y = pd.get_dummies(df['label'])
y = y.iloc[:,0].values
```

```python
y.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features = 100,ngram_range=(2,2))
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size=0.25, random_state=42)
X_train = tfidf.fit_transform(X_train).toarray()
```

```python
X_test = tfidf.transform(X_test).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems = 30,linewidth = 100000,formatter = dict(float = lambda x:"%.3g" %x))
X_train
```

```python
tfidf.vocabulary_
```

### Machine Learning Model Training (Multinomial Naive Bayes Algorithm)

```python
from sklearn.naive_bayes import MultinomialNB
spam_tfidf_model = MultinomialNB().fit(X_train,y_train)
```

```python
## Predict
y_pred = spam_tfidf_model.predict(X_test)
```

```python
# Predict Result
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
