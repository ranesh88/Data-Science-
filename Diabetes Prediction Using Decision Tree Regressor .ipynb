### Diabetes Predictions

```python
from sklearn.datasets import load_diabetes
dataset = load_diabetes()
dataset
```

```python
print(dataset['DESCR'])
```

```python
import pandas as pd
df_diabetes = pd.DataFrame(dataset.data,columns = [
    'age',
    'sex',
    'bmi',
    'bp',
    'tc',
    'ldl',
    'hdl',
    'tch',
    'ltg',
    'glu'
])
```

```python
df_diabetes
```

```python
X = df_diabetes ## Independent features
y = dataset['target']  ## Dependent feature
```

```python
## train-test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33,random_state = 42)
```

```python
## Corrrelation
X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (15,10))
sns.heatmap(X_train.corr(),annot = True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
score = r2_score(y_pred,y_test)
print(score)
print(mean_absolute_error(y_pred,y_test))
print(mean_squared_error(y_pred,y_test))
```

## Hyperparameter Tuning with GridsearchCV

```python
param_grid = {
    'criterion' :['squared_error','friedman_mse','absolute_error'],
    'splitter' :['best','random'],
    'max_depth' :[1,2,3,4,5,10,15,20,25],
    'max_features' :['auto','sqrt','log2']
}
```

```python
param_grid
```

```python
from sklearn.model_selection import GridSearchCV
regressor = DecisionTreeRegressor()
grid = GridSearchCV(regressor,
                   param_grid = param_grid,
                   cv = 5,
                   scoring = 'neg_mean_squared_error')
```

```python
grid
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
## Predictions
y_pred = grid.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
score = r2_score(y_pred,y_test)
print(score)
print(mean_absolute_error(y_pred,y_test))
print(mean_squared_error(y_pred,y_test))
```

```python
selectmodel = DecisionTreeRegressor(criterion = 'friedman_mse',
                                   max_depth = 3,
                                   max_features = 'sqrt',
                                   splitter = 'best')
```

```python
selectmodel.fit(X_train,y_train)
```

```python
## Visualize the Decision Tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(selectmodel,filled = True)
```
