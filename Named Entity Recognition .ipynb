### Named Entity Recognition

```python
paragraphs ="""
The Eiffel Tower, located in Paris, France, is one of the most iconic landmarks in the world and a symbol of French culture and innovation. Standing at a height of 330 meters (1,083 feet) after the addition of antennas in 2022, it was originally constructed between 1887 and 1889 for the 1889 World's Fair to celebrate the centennial of the French Revolution. Designed by Gustave Eiffel and his engineering team, the tower is made of wrought iron and features three levels accessible to visitors, with restaurants on the first two levels and a panoramic observation deck at the top. Initially met with criticism from artists and intellectuals, the tower eventually gained widespread admiration and held the title of the tallest man-made structure in the world until 1930. Today, it attracts nearly 7 million visitors annually, making it one of the most visited monuments globally. A masterpiece of engineering, the Eiffel Tower expands and contracts by up to six inches depending on the temperature and has been repainted numerous times, transitioning from its original red to the current Eiffel Tower Brown. Its rich history, stunning views, and cultural significance continue to captivate visitors from around the world.
"""
```

```python
print(paragraphs)
```

```python
import nltk
word_list = nltk.word_tokenize(paragraphs)
```

```python
print(word_list)
```

```python
pos_tag = nltk.pos_tag(word_list)
```

```python
pos_tag
```

```python
# Perform Named Entity Recognition
nltk.download('maxent_ne_chunker')
nltk.download('words')
ner_tree = nltk.ne_chunk(pos_tag)

# Print the tree structure
print(ner_tree)
```

```python
from gensim.models import Word2Vec

# Sample data: list of tokenized sentences
sentences = [["hello", "world"], ["machine", "learning", "is", "fun"]]

# Train a Word2Vec model
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)

# Access word vectors
vector = model.wv['hello']
print(vector)
```

```python
from gensim.models import Word2Vec

# Sample dataset: tokenized sentences
sentences = [
    ["artificial", "intelligence", "is", "transforming", "industries"],
    ["machine", "learning", "is", "a", "subset", "of", "artificial", "intelligence"],
    ["deep", "learning", "is", "a", "type", "of", "machine", "learning"],
    ["natural", "language", "processing", "is", "part", "of", "artificial", "intelligence"]
]

# Train the Word2Vec model
model = Word2Vec(
    sentences,  # input dataset
    vector_size=50,  # dimensionality of word vectors
    window=3,        # context window size
    min_count=1,     # minimum word frequency
    workers=4,       # number of worker threads
    sg=1             # skip-gram (set to 0 for CBOW)
)

# Access word vectors
word_vector = model.wv['artificial']  # Get the vector for the word 'artificial'
print(f"Vector for 'artificial':\n{word_vector}")

# Most similar words
similar_words = model.wv.most_similar('intelligence', topn=3)
print(f"Most similar words to 'intelligence':\n{similar_words}")
```

```python
from gensim.test.utils import common_texts
from gensim.models import Word2Vec

# Common_texts is a small example dataset included with Gensim
model = Word2Vec(common_texts, vector_size=50, window=5, min_count=1, workers=4)

# Get a vector for a word
vector = model.wv['computer']
print(f"Vector for 'computer':\n{vector}")

# Most similar words
similar_words = model.wv.most_similar('computer', topn=3)
print(f"Most similar words to 'computer':\n{similar_words}")
```
