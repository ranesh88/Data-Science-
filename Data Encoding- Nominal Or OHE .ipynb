```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## Create a simple DataFrame
df = pd.DataFrame ({
    'color':['red','green','blue','green','red','blue','green','red','red']
})
```

```python
df
```

```python
## Create an instance of onehotEncoder
encoder = OneHotEncoder()
```

```python
## Perform fit and transform
encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
encoder_df = pd.DataFrame(encoded,columns = encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
## for new data
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis = 1)
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips[['sex','smoker','day','time']].value_counts()
```

```python
## create an instance of OneHotEncoder
tips_encoder = OneHotEncoder()
```

```python
encoded_tips = tips_encoder.fit_transform(tips[['sex','smoker','day','time']]).toarray()
```

```python
encoded_tips
```

```python
tips_df = pd.DataFrame(encoded_tips,columns = tips_encoder.get_feature_names_out())
```

```python
tips_df
```

```python
pd.concat([tips,tips_df],axis=1)
```
